```python
#Outlier Analysis
#q1
def outliers_iqr(data, column_name):
    q1 = data[column_name].quantile(0.25)
    q3 = data[column_name].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    return outliers

me_outlier_score = outliers_iqr(data_middle_east, 'score')
sa_outlier_score = outliers_iqr(data_south_asia, 'score')
me_outlier_gdp = outliers_iqr(data_middle_east, 'Log GDP per capita')
sa_outlier_gdp = outliers_iqr(data_south_asia, 'Log GDP per capita')
```

```python
sa_concat_oulier = pd.concat([sa_outlier_score, sa_outlier_gdp]).drop_duplicates()
me_concat_oulier = pd.concat([me_outlier_score, me_outlier_gdp]).drop_duplicates()
```

```python
plt.scatter(
    data_middle_east['Log GDP per capita'],
    data_middle_east['score'],
    label='Middle East Data',
)
plt.scatter(
    me_concat_oulier['Log GDP per capita'],
    me_concat_oulier['score'],
    label='Middle East Outlier',
)
plt.xlabel('GDP per capita')
plt.ylabel('Happiness Score')
plt.title('Outliers middle east')
plt.legend()
plt.show()
```

```python
plt.scatter(
    data_south_asia['Log GDP per capita'],
    data_south_asia['score'],
    label='South Asia',
    color='blue'
)
plt.scatter(
    sa_concat_oulier['Log GDP per capita'],
    sa_concat_oulier['score'],
    label='South Asia Outlier',
    color='red'
)
plt.xlabel('GDP per capita')
plt.ylabel('Happiness Score')
plt.title('Outliers south asia')
plt.legend(loc='upper left')
plt.show()
```
